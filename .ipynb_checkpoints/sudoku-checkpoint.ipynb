## Part One
You should write all of your code for solving sudokus below this cell.

You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
import random
import copy

import numpy as np


class SudokuState:
    """Representation of sodoku board"""

    # TEST A DEEP COPY ALTERNATIVE BY USING DEFAULT ARGS

    def __init__(self, sudoku, possible=None):  # using default arguments to pass on
        self.final_values = sudoku  # 2d np array
        self.mc_len = 9  # most contrained cells len
        self.mc_row, self.mc_col = -1, -1  # most contrained cells row, col
        if possible is None:
            self.possible_values = [[[i for i in range(1, 10)] for _ in range(0, 9)] for _ in range(0, 9)]
        else:
            self.possible_values = possible

    def initial_propagate(self):
        # loops through and reduces the possible moves down through checking the contraints
        for row in range(0, 9):
            for col in range(0, 9):
                if self.final_values[row, col] != 0:  # if the values have already been placed propagate them
                    if self.check_board(row, col):
                        self.propagate(row, col)
                    else:
                        return False
                else:  # if there is no value in this position then no need to propagate %%!
                    continue
        return True

    def check_board(self, row, col):
        # for a given cell in the grid
        # check whether it appears in the column
        # check whether it appears in the row
        # check whether it appears in the box

        number = self.final_values[row, col]
        for i in range(0, 9):
            if self.final_values[row, i] == number and i != col:  # checks the horizontal
                return False
            if self.final_values[i, col] == number and i != row:  # checks the vertical
                return False
        # checks the box
        row0 = (row // 3) * 3
        col0 = (col // 3) * 3
        for i in range(0, 3):
            for j in range(0, 3):
                if number == self.final_values[row0 + i, col0 + j] and (row0 + i != row and col0 + i != col):
                    return False
        return True

    def get_final_state(self):
        if self.is_goal():
            return self.final_values
        else:
            return -1

    def is_invalid(self):
        """This partial state is invalid if any column has no possible values"""
        for row in range(0, 9):
            for col in range(0, 9):
                if self.final_values[row, col] == 0 and len(self.possible_values[row][col]) == 0:
                    return True
        return False

    # return any(len(values) == 0 for values in self.possible_values)

    def is_goal(self):
        """This partial state is a goal state if every column has a final value"""
        # return all(cell != 0 for cell in row for row in self.final_values)
        for row in range(0, 9):
            for col in range(0, 9):
                if self.final_values[row, col] == 0:
                    return False
        return True
        # return all(all(row != 0) for row in self.final_values)

    def get_singleton_cells(self):
        """Returns the columns which have no final value but exactly 1 possible value"""
        singletons = []
        for row in range(0, 9):
            for col in range(0, 9):
                if len(self.possible_values[row][col]) == 1 and self.final_values[row, col] == 0:
                    singletons.append([row, col])
        return singletons

    def propagate(self, row, col):  # CAN BE OPTIMISED THROUGH CHECKING ONE SIDE THEN THE OTHER - similar to kweens
        # check number in the [row][col] -done
        # then going to change that number in possible_values as it has already been decided -done
        # then going to propagate through the horizontal possible_values, removing the value at [row][col]
        # same with the vertical
        # again with the box
        # done
        # if the number already exists in the row/col/box returning false
        number = self.final_values[row, col]
        self.possible_values[row][col] = []  # setting the singleton
        for i in range(0, 9):
            if number in self.possible_values[row][i]:  # checks the horizontal
                self.possible_values[row][i].remove(number)
            if number in self.possible_values[i][col]:  # checks the vertical
                self.possible_values[i][col].remove(number)
        # checks the box
        row0 = (row // 3) * 3
        col0 = (col // 3) * 3
        for i in range(0, 3):
            for j in range(0, 3):
                if number in self.possible_values[row0 + i][col0 + j]:
                    self.possible_values[row0 + i][col0 + j].remove(number)


    """"
    def check_valid_move(self, row, col, val):
        for i in range(0, 9):
            if val in self.final_values[row][i]:  # checks the horizontal
                self.possible_values[row][i].remove(val)
            if val in self.final_values[i][col]:  # checks the vertical
                self.possible_values[i][col].remove(val)
        # checks the box
        row0 = (row // 3) * 3
        col0 = (col // 3) * 3
        for i in range(0, 3):
            for j in range(0, 3):
                if val in self.final_values[row0 + i][col0 + j]:
                    self.possible_values[row0 + i][col0 + j].remove(val)
    """

    def create_possible_copy(self):
        possible_copy = [[[] for _ in range(0, 9)] for _ in range(0, 9)]
        for row in range(0, 9):
            for col in range(0, 9):
                for values in self.possible_values[row][col]:
                    possible_copy[row][col].append(values)
        return possible_copy

    def set_value(self, row, col, val):
        # checks that the value can be placed there
        # creates new state and updates its final_values then propagates it through possible_values
        # then is there are any singletons left in possible values, it automatically places them and propagates

        # checking the value can be placed there
        if val not in self.possible_values[row][col]:
            raise ValueError(f"{val} is not a valid choice for position: {row},{col}")

        # create a deep copy: the method returns a new state, does not modify the existing one
        # then updates its final values and propagates

        possible_copy = self.create_possible_copy()
        final_copy = self.final_values.copy()
        state = SudokuState(final_copy, possible_copy)

        state.final_values[row, col] = val
        #print("POSSIBLE VAL", state.possible_values[row][col])
        state.propagate(row, col)

        singletons = state.get_singleton_cells()
        # print("SINGLETONS: ",singletons)
        while len(singletons) > 0:
            col = singletons[0][1]
            row = singletons[0][0]
            # print(f"col: {col} row: {row}")
            state = state.set_value(row, col, state.possible_values[row][col][0])
            singletons = state.get_singleton_cells()

        return state


def pick_next_cell(state):
    # cell_indices = [index for index, values in enumerate(state.possible_values) if len(values) > 1]
    # print("CELL INDICES" , cell_indices)
    for row in range(0, 9):
        for col in range(0, 9):
            if state.final_values[row][col] == 0:
                # print("ROW: ",row, "COL: ", col)
                return row, col
    return -1, -1

def pick_next_cell1(state):
    min = 9
    for row in range(0, 9):
        for col in range(0, 9):
            if 1 < len(state.possible_values[row][col]) <= min:
                min_r = row
                min_c = col
                min = len(state.possible_values[row][col])
            return min_r, min_c



def depth_first_search(state):
    row, col = pick_next_cell(state)
    #row = state.mc_row
    #col = state.mc_col
    values = state.possible_values[row][col]
    # print("Called DFS")

    for value in values:
        new_state = state.set_value(row, col, value)
        if new_state.is_goal():
            return new_state
        if not new_state.is_invalid():
            deep_state = depth_first_search(new_state)
            if deep_state is not None and deep_state.is_goal():
                return deep_state

    return None





def sudoku_solver(sudoku):
    """
    Solves a Sudoku puzzle and returns its unique solution.

    Input
        sudoku : 9x9 numpy array
            Empty cells are designated by 0.

    Output
        9x9 numpy array of integers
            It contains the solution, if there is one. If there is no solution, all array entries should be -1.


    sudoku_state = SudokuState(sudoku)
    sudoku_state.initial_propagate()
    print("POSSIBLE MOVES: ",sudoku_state.possible_values)
    print(sudoku_state.get_singleton_cells())
    print(sudoku_state.final_values)
    print(sudoku_state.possible_values[2][7][0])
    sudoku_state.set_value(2,7,7)
    print(sudoku)
    """

    partial_state = SudokuState(sudoku)
    if partial_state.initial_propagate() == False:
        return (np.full((9, 9), -1))
    solved = depth_first_search(partial_state)
    # print(solved.final_values)
    if solved == None:
        return (np.full((9, 9), -1))

    return solved.final_values

# start_time = time.time()
# partial_state = SudokuState(sudoku)
# goal = depth_first_search(partial_state).get_final_state()
# print(goal)
# print("--- %s seconds ---" % (time.time() - start_time))
# """


### YOUR CODE HERE

# return solved_sudoku
```
