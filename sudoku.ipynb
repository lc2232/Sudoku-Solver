## Part One
You should write all of your code for solving sudokus below this cell.

You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
import numpy as np


class SudokuState:
    """Representation of Sudoku board and methods to solve it and check whether it can be solved"""

    def __init__(self, sudoku, possible=None):
        """
        Constructor for the SudokoState

        :param sudoku: The 2D numpy array represeting the Sudoku grid
        :param possible: Default - None, when no prior state has been used
        """

        self.final_values = sudoku  # 2d np array
        self.mc_len = 9  # most constrained cells len
        self.mc_row, self.mc_col = 0, 0  # most constrained cells row, col
        if possible is None:
            self.possible_values = [[i for i in range(0, 9)] for _ in range(0, 9)]
            self.build_possible_values()
        else:
            self.possible_values = possible

    def build_possible_values(self):
        for row in range(0,9):
            for col in range(0,9):
                if self.final_values[row,col] != 0:
                    self.possible_values[row][col] = []
                else:
                    self.possible_values[row][col] = [i for i in range(1, 10)]


    def initial_propagate(self):
        """
        Iterates through the grid checking whether it can be solved and propogated the values placed there already

        :return: True if it can be solved. False if not
        """

        for row in range(0, 9):
            for col in range(0, 9):
                if self.final_values[row, col] != 0:  # if the values have already been placed propagate them
                    if self.check_board(row, col):
                        self.propagate(row, col)
                    else:
                        return False
                else:  # if there is no value in this position then no need to propagate %%!
                    continue
        return True

    def check_board(self, row, col):
        """
        Checks whether the board is solvable i.e. does it go against any of the constraints off the bat

        :param row: row index of a cell to be checked
        :param col: col index of a cell to be checked
        :return: True if solvable False if not
        """

        number = self.final_values[row, col]
        for i in range(0, 9):
            if self.final_values[row, i] == number and i != col:  # checks the horizontal
                return False
            if self.final_values[i, col] == number and i != row:  # checks the vertical
                return False
        # checks the box
        row0 = (row // 3) * 3
        col0 = (col // 3) * 3
        for i in range(0, 3):
            for j in range(0, 3):
                if number == self.final_values[row0 + i, col0 + j] and (row0 + i != row and col0 + i != col):
                    return False
        return True

    def is_invalid(self):
        """This partial state is invalid if any cell has the value 0 and has 0 possible moves left"""

        for row in range(0, 9):
            for col in range(0, 9):
                if self.final_values[row, col] == 0 and len(self.possible_values[row][col]) == 0:
                    return True
        return False

    def is_goal(self):
        """This compares each cell value with 0 - meaning it is incomplete"""

        for row in range(0, 9):
            for col in range(0, 9):
                if self.final_values[row, col] == 0:
                    return False
        return True
        # return all(all(row != 0) for row in self.final_values)

    def get_singleton_cells(self):
        """Returns the cells which have no final value but exactly 1 possible value"""

        singletons = []
        for row in range(0, 9):
            for col in range(0, 9):
                if len(self.possible_values[row][col]) == 1 and self.final_values[row, col] == 0:
                    singletons.append([row, col])
        return singletons

    def propagate(self, row, col):
        """
        Propagates a particular row and columns value throughout its domain, and changes
        the most constrained cells values.

        :param row: row of cell being propagated
        :param col: col of cell being propagated
        """

        number = self.final_values[row, col]
        self.possible_values[row][col] = []  # removing any possible values

        for i in range(0, 9):
            if number in self.possible_values[row][i]:  # checks the horizontal
                self.possible_values[row][i].remove(number)
                if 1 < len(self.possible_values[row][i]) < self.mc_len:
                    self.mc_len = len(self.possible_values[row][i])
                    self.mc_col = i
                    self.mc_row = row

            if number in self.possible_values[i][col]:  # checks the vertical
                self.possible_values[i][col].remove(number)
                if 1 < len(self.possible_values[i][col]) < self.mc_len:
                    self.mc_len = len(self.possible_values[i][col])
                    self.mc_col = col
                    self.mc_row = i
        # checks the box
        row0 = (row // 3) * 3
        col0 = (col // 3) * 3
        for i in range(0, 3):
            for j in range(0, 3):
                if number in self.possible_values[row0 + i][col0 + j]:
                    self.possible_values[row0 + i][col0 + j].remove(number)
                    if 1 < len(self.possible_values[row0 + i][col0 + j]) < self.mc_len:
                        self.mc_len = len(self.possible_values[row0 + i][col0 + j])
                        self.mc_col = col0 + j
                        self.mc_row = row0 + i

    def create_possible_copy(self):
        """
        Alternative method to deepcopy which takes up a large amount of computational time, creates a copy of the
        possible_values 3D list.

        :return: copy of possible_values
        """
        possible_copy = [[[] for _ in range(0, 9)] for _ in range(0, 9)]
        for row in range(0, 9):
            for col in range(0, 9):
                for values in self.possible_values[row][col]:
                    possible_copy[row][col].append(values)
        return possible_copy

    def set_value(self, row, col, val):

        """
        Checks that the value can be placed in the given row and col then creates new state and updates its
        final_values then propagates this change through possible_values. If there are any singletons left then
        fills them in and propagates them.

        :param row: The row index where the value should be placed
        :param col: The column index where the value should be placed
        :param val: The value that should be placed at the row/col index
        :return: state: the new state created
        """

        # checking the value can be placed there and raising an error if not a valid move
        if val not in self.possible_values[row][col]:
            raise ValueError(f"{val} is not a valid choice for position: {row},{col}")

        # creates copies of possible moves and final values and creates new state
        possible_copy = self.create_possible_copy()
        final_copy = self.final_values.copy()
        state = SudokuState(final_copy, possible_copy)

        # assigns value to the new state
        state.final_values[row, col] = val
        state.propagate(row, col)

        singletons = state.get_singleton_cells()
        while len(singletons) > 0:
            col = singletons[0][1]
            row = singletons[0][0]

            state.final_values[row, col] = state.possible_values[row][col][0]
            state.propagate(row, col)
            singletons = state.get_singleton_cells()

        return state


def find_first_empty(state):
    """
    Finds the first empty cell in-case the Sudoku grid if the most constrained values
    do not pass the guard

    :param state: An unsolved Sudoku State
    :return: row, col: the indexes of the first unsolved value
    """
    for row in range(0, 9):
        for col in range(0, 9):
            if state.final_values[row][col] == 0:
                return row, col
    return -1, -1


def depth_first_search(state):
    """
    Performs a DFS on a SudokuState to solve it

    :param state: A SudokuState that requires solving
    :return: The solved SudokuState or None depending on whether it is solvable or not
    """
    # Guard for most constrained values
    if state.final_values[state.mc_row, state.mc_col] == 0:
        row = state.mc_row
        col = state.mc_col
    else:
        row, col = find_first_empty(state)

    values = state.possible_values[row][col]

    for value in values:
        new_state = state.set_value(row, col, value)
        if new_state.is_goal():
            return new_state
        if not new_state.is_invalid():
            deep_state = depth_first_search(new_state)
            if deep_state is not None and deep_state.is_goal():
                return deep_state

    return None


def sudoku_solver(sudoku):
    """
    Solves a Sudoku puzzle and returns its unique solution.

    Input
        sudoku : 9x9 numpy array
            Empty cells are designated by 0.

    Output
        9x9 numpy array of integers
            It contains the solution, if there is one. If there is no solution, all array entries should be -1.
    """

    partial_state = SudokuState(sudoku)
    if not partial_state.initial_propagate():
        return np.full((9, 9), -1)
    solved = depth_first_search(partial_state)
    if solved is None:
        return np.full((9, 9), -1)

    return solved.final_values
```
